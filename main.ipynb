```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.30)
```

```python
# standardization- feature scaling techinique
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
# Elbow Method
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# plot the Elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python
kmeans.fit_predict(X_train_scaled)
```

```python
y_pred=kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

```python
# silhoutee score
from sklearn.metrics import silhouette_score
```

```python
silhoutee_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    scores=silhouette_score(X_train_scaled,kmeans.labels_)
    silhoutee_coefficients.append(scores)
```

```python
silhoutee_coefficients
```

```python
# plot the Elbow curve
plt.plot(range(2,11),silhoutee_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhoutee_coefficients")
plt.show()
```
